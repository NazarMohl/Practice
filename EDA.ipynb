```python
year = 2020
month = 2
energy_source = 'Coal (Short Tons)'
df_total_producer_md.loc[\
                         (df_total_producer_md['YEAR'] == year) \
                         & (df_total_producer_md['MONTH'] == month) \
                         & (df_total_producer_md['ENERGY SOURCE'] == energy_source)\
                        ].to_dict()
```

```python
years = df_total_producer_md['YEAR'].unique()
energy_source = 'Coal (Short Tons)'

for year in (years[:-1]):
    yearly_consumption = 0
    for month in range(1,13):
        monthly_consumption = df_total_producer_md.loc[\
                                                       (df_total_producer_md['YEAR'] == year) \
                                                       & (df_total_producer_md['MONTH'] == month) \
                                                       & (df_total_producer_md['ENERGY SOURCE'] == energy_source)\
                                                      ]['CONSUMPTION'].sum()
        yearly_consumption = yearly_consumption + monthly_consumption
        print("For ", year, "/",month, "Energy Source: ", energy_source, "Monthly Consumption = ", monthly_consumption)
    print("yearly Consumption", yearly_consumption)
#     df_yearly_producer_md = df_total_producer_md.loc[\
#                                                      (df_total_producer_md['YEAR'] == year) \
#                                                      & (df_total_producer_md['ENERGY SOURCE'] == 'Coal (Short Tons)')\
#                                                     ]['CONSUMPTION']
```
