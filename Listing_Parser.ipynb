```python
def list_cleaner(line):
    if line[0:7] == ' Volume':
        return
    elif line[24:40] == '<DIR>          .':
        return
    elif line[17:21] == 'File':
        return
    return line
```

```python
dir_heads = []
for line in listing:
    clean_line = list_cleaner(line)
    if clean_line != None:
        dir_heads.append(line)
        clean_line = list_cleaner(line)
dir_heads
```

```python
listing = dir_heads
```

```python
dir_heads = []
for line in listing:
#     count = line.count('\\')
#     if count > 7:
    if line[6:10] == '2022':
        line = line[39:]
    if line.count('.git') > 0:
        continue
    if line.count('.idx') > 0:
        continue
    if line.count('\\') >= 7:
        continue
    if line.count('.md') > 0:
        continue
    if line.count('.canvas') > 0:
        continue
    if line.count('.learn') > 0:
        continue
    if line.count('.png') > 0:
        continue
    if line.count('.gz') > 0:
        continue
    if line.count('.zip') > 0:
        continue
    if line.count('.csv') > 0:
        continue
    if line.count('.sample') > 0:
        continue
    if line.count('.jpg') > 0:
        continue
    if line.count('.JPG') > 0:
        continue
    if line.count('.jpeg') > 0:
        continue
    if line.count('.gif') > 0:
        continue
    if line.count('.pdf') > 0:
        continue
    if line.count('.xml') > 0:
        continue
    if line.count('.txt') > 0:
        continue
    if line.count('.mp4') > 0:
        continue
    if line.count('.pack') > 0:
        continue
    if line.count('.js') > 0:
        continue
    if line.count('bytes free') > 0:
        continue
    if line.count('\\images') > 0:
        continue
    if line.count('\\index_file') > 0:
        continue
    if line.count('\\data') > 0:
        continue
    if (("0" in line) or ("1"in line) or ("2" in line)):
        dir_heads.append(line)
dir_heads
```
